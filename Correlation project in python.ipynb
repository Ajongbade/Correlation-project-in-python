```python
# import libraries 


import pandas as pd 
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt 
plt.style.use ('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline 
matplotlib.rcParams ['figure.figsize'] = (12,8) 

```

```python
#read data
df= pd.read_csv(r'C:\Users\dele\Desktop\certificates\Movies Project\movies.csv')
```

```python
df.head()
```

```python
# verify all cells are filled
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
# drop missing cells 

df = df.dropna(how = 'any', axis = 0)
```

```python
#verify 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
#data types 

df.dtypes
```

```python
#change buget, votes, gross & runtime 

df['budget'] = df['budget'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
```

```python
df.head()
```

```python
#sort table by gross income

df.sort_values(by = ['gross'], inplace = False, ascending = False).head()
```

```python
#allow entire dataset in view

pd.set_option('display.max_rows', None)
```

```python
df.head()
```

```python
# drop duplicates

df['company'].drop_duplicates().sort_values(ascending = False).head()
```

```python
#scatter plot budget vs gross revenue

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('budget vs gross')
plt.xlabel('budget')
plt.ylabel('gross')

plt.show()
```

```python
#plot budget vs gross using seaborn

sns.regplot(x='budget', y = 'gross', data=df, scatter_kws = {"color": "red"}, line_kws = {"color":"blue"})
```
